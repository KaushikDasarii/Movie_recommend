```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movie['genres']= movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords']= movie['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    counter = 0

    if not isinstance(obj, str) or not obj.strip().startswith("["):
        return L  # skip non-strings or irrelevant data

    try:
        # Attempt to fix common issues like missing closing bracket
        if obj.strip()[-1] != ']':
            obj = obj.strip() + ']'

        parsed = ast.literal_eval(obj)
        for i in parsed:
            if counter < 3 and isinstance(i, dict) and 'name' in i:
                L.append(i['name'])
                counter += 1
    except Exception as e:
        print(f"Error parsing: {obj} -- {e}")
        # You can log the error or skip
        return L

    return L
```

```python
movie['cast']= movie['cast'].apply(convert3)
```

```python
movie
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
           L.append(i['name'])
           break
    return L
```

```python
movie['crew']= movie['crew'].apply(fetch_director)
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie['tags']= movie['overview']+movie['cast']+movie['keywords']+movie['crew']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append( ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    
    # Sort by similarity score (excluding the movie itself)
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    recommended_movies = []
    for i in movie_list:
        recommended_movies.append(new_df.iloc[i[0]].title)
    
    return recommended_movies
```

```python
print("Recommended for 'Inception':")
print(recommend("Inception"))
```

```python
print("\nRecommended for 'The Matrix':")
print(recommend("The Matrix"))
```
